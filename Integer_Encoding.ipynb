```python
import pandas as pd

# to split the datasets
from sklearn.model_selection import train_test_split
```

```python
# Create mapping dictionary of unique values in column
def find_category_mappings(df, var):
    return {k: i for i, k in enumerate(df[var].unique(), 0)}

df = pd.read_csv("penguins_size.csv")
is_cat = ["species", "island", "sex"]
df = df.dropna()

# Dataframe of raw dataset. Needs encoding to be processed
display(df.head())

# Split the dataframe into testing and training datasets
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(labels=["species"], axis=1),  # predictors
    df["species"],  # target
    test_size=0.3,  # percentage of obs in test set
    random_state=0)  # seed to ensure reproducibility

print(X_train.shape, X_test.shape)
display(X_train.head())

# Encode using our function
for col in is_cat:
    # ignore species column as it's our labels
    if col != "species":
        ordinal_mapping = find_category_mappings(df, col)
        X_train[col] = X_train[col].map(ordinal_mapping)
        X_test[col] = X_test[col].map(ordinal_mapping)

display(X_train.head())

```
